```python
df['salaire'] = df['salaire'].astype(str)

df['salaire'] = df['salaire'].str.replace('{','').str.replace('}','')
df['salaire'] = df['salaire'].str.replace("''", "")
```

```python
import pandas as pd
import re

for index, row in df.iterrows():
    salaire = row['salaire']
    # Extraire le premier nombre dans la chaîne
    match = re.search(r'\d+', salaire)
    if match:
     nombre = int(match.group())
     df.at[index, 'salaire'] = nombre

print(df['salaire'])
```

```python
#pour obtenir le salaire par moi quand on récupère que le salaire par heure 
for index, row in df.iterrows():
    salaire = row['salaire']
    if salaire < 200:
        df.at[index, 'salaire'] = salaire * 35*4
```
