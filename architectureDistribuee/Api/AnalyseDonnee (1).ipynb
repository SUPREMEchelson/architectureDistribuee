```python
import pandas as pd
import re


# Parcourir chaque ligne du DataFrame
for index, row in df.iterrows():
    salaire = row['salaire']
    # Extraire le premier nombre dans la chaîne
    match = re.search(r'\d+', salaire)
    if match:
     nombre = int(match.group())
     df.at[index, 'salaire'] = nombre

# Afficher le DataFrame avec la colonne "mensuel" mise à jour
print(df['salaire'])
```

```python
# Parcourir chaque ligne du DataFrame
for index, row in df.iterrows():
    salaire = row['salaire']
    # Vérifier si la valeur est inférieure à 200
    if salaire < 200:
        df.at[index, 'salaire'] = salaire * 35*4
```

```python
def correlation_emploi_loyer(df):
    emploi_par_ville = df.groupby('Ville').size()
    prix_moyen_par_ville = df.groupby('Ville')['Prix par mois'].mean()
    corr = emploi_par_ville.corr(prix_moyen_par_ville)
    return corr
```

```python
print(correlation_emploi_loyer(df))
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Group the data by city and count the number of job offers per city
job_counts = df.groupby('Ville')['intitule'].count()

# Group the data by city and calculate the correlation between job offers and rent price per city
corr_per_city = df.groupby('Ville')[['intitule', 'Prix par mois']].corr().iloc[0::2,-1]

# Plot the results as a bar chart and scatter plot
fig, ax = plt.subplots(2, 1, figsize=(8, 10))
job_counts.plot(kind='bar', ax=ax[0])
ax[0].set_title("Nombre d'offres d'emploi par ville")
ax[0].set_xlabel("Ville")
ax[0].set_ylabel("Nombre d'offres d'emploi")
corr_per_city.plot(kind='bar', ax=ax[1])
ax[1].set_title("Corrélation entre le nombre d'offres d'emploi et le prix du loyer par ville")
ax[1].set_xlabel("Ville")
ax[1].set_ylabel("Corrélation")
plt.tight_layout()
plt.show()

```
